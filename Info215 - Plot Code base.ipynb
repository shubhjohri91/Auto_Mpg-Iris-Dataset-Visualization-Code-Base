```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(style="white", color_codes=True)
iris = pd.read_csv("Iris.csv")
```

# Scatter Plot

```python
sns.FacetGrid(iris,hue='Species',size=5)\
.map(plt.scatter,'SepalLengthCm','SepalWidthCm')\
.add_legend()
plt.title('Scatterplot of Sepal Length vs Sepal Width', fontsize = large);
```

# BoxPlot

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.boxplot(x='Species',y='PetalLengthCm',data=iris,order=['Iris-virginica','Iris-versicolor','Iris-setosa'],linewidth=2.5,orient='v',dodge=False)
plt.title('Boxplot of Petal length by species',fontsize=large);
```

# Strip plot

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.stripplot(x='Species',y='SepalLengthCm',data=iris,jitter=True,edgecolor='gray',size=8,palette='winter',orient='v')
plt.title('Strip plot of Sepal length by species',fontsize=large);
```

# Density Plot

```python
sns.FacetGrid(iris, hue="Species", size=6) \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend();
plt.title('Density plot of Petal Lenth by Species', fontsize = large);
plt.ylabel('Density');
```

# Pair Plot

```python
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3);
```

# LM Plot

```python
fig=sns.lmplot(x="PetalLengthCm", y="PetalWidthCm",data=iris)
plt.title('Petal Lenth vs Petal Width', fontsize = large);
```

# Bar Plot

```python
sns.countplot('Species',data=iris)
plt.title('Box Plot of Species counts', fontsize = large);
plt.show()
```

# Pie Chart

```python
plt.figure(figsize=(10,8))
plt.pie(iris['Species'].value_counts(),labels =["Iris-setosa","Iris-vericolor","Iris-virginica"],explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True)
plt.title('Pie chart of Species counts', fontsize = large)
plt.show();
```

# Violin Plot

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.violinplot(x='Species',y='SepalLengthCm',data=iris)
plt.title('Violin plot of Sepal Lenth by Species', fontsize = large);
```

# Heat map

```python
plt.figure(figsize=(10,7))
sns.heatmap(iris.corr(),annot=True,cmap='summer')
plt.title("Heatmap", fontsize = large);
```
